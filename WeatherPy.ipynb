```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Import API key
import api_keys

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

## Perform API Calls

```python
# OpenWeatherMap API Key
api_key = api_keys.api_key

# Starting URL for Weather Map API Call
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + api_key 

lat = []
lon = []
temp = []
humi = []
cloud = []
wind = []
city_list = []

for city in cities:
    query_url = url + "&q=" + city
    response = requests.get(query_url)
    
    if response.status_code == 404:
        next    
    else:
        print(f"Processing Record {cities.index(city)+1} | {city}")
        print(query_url)
        
        response_json = response.json()
        
        city_list.append(city)
        temp.append(response_json["main"]["temp_max"])
        humi.append(response_json["main"]["humidity"])
        cloud.append(response_json["clouds"]["all"])
        wind.append(response_json["wind"]["speed"])
        lat.append(response_json["coord"]["lat"])
        lon.append(response_json["coord"]["lon"])
```

```python
city_df = pd.DataFrame({"City": city_list, 
                        "Cloudiness": cloud, 
                        "Humidity": humi, 
                        "Max Temp": temp, 
                        "Wind Speed": wind, 
                        "Lat": lat, 
                        "Long": lon
                       })
```

```python
LatVsTemp = pd.DataFrame({"City": city_list, "Lat": lat, "Temp": temp})
LatVsTemp = LatVsTemp.set_index("City")

plt.scatter(lat, temp, marker="o", facecolor="red", edgecolor="black")

plt.ylim(0, 120, 1)
plt.xlim(-70, 90, 1)

plt.title("City Latitude VS. Max Temperature (07/22/18)")
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (F)")

plt.grid()
plt.savefig('LatVsTemp.png')
```
